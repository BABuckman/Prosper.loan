## Investigation Overview

> This presentation presents some findings in our analysis of the prosper loan dataset. It will investigate the relation between loans taken (original loan amount), loan terms, borrower APR, borrower Rate, and Prosper Loan Ratings (Alpha) \
> Key findings are: 
> - The BorrowerAPR decrease with the increasing loan terms for people with HR through to C ratings. But this trend changes from prosperRating (Alpha) B through AA with longer loan terms (60 months) being higher. 
> - Current loan status category is the highest loan status category in the dataset. 
> - The BorrowerAPR decrease with the increasing loan terms for people with HR through to C ratings. 


## Dataset Overview

> This data set contains prosper loan information for 113937 entries and 16 features. 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_loan = pd.read_csv('df_loan.csv')
```

## Visualization of BorrowerAPR and ProsperRating (Alpha)

> This shows that the BorrowerAPR and BorrowerRate decreases with increase ProsperRating (Alpha). That is, borrowers with improved ratings have the lowest APR. \
> BorrowerAPR and BorrowerRate affects the Prosper Rating (Alpha)

```python
# Convert ProsperRating (Alpha) into ordered category

order = ['HR','E','D','C','B','A','AA']
ordered_alpha = pd.api.types.CategoricalDtype(ordered = True, categories = order)
df_loan['ProsperRating (Alpha)'] = df_loan['ProsperRating (Alpha)'].astype(ordered_alpha)
```

```python
base_color = sb.color_palette()[0]

plt.figure(figsize = [10,6])
plt.subplot(1,2,1)
sb.boxplot(data = df_loan, x = 'ProsperRating (Alpha)', y = 'BorrowerAPR', color = base_color)
plt.xlabel('Prosper Rating (Alpha)')
plt.ylabel('Borrower APR')
plt.title('Distribution of Borrower APR and Prosper Rating')

plt.subplot(1,2,2)
sb.boxplot(data = df_loan, x = 'ProsperRating (Alpha)', y = 'BorrowerRate', color = base_color)
plt.xlabel('Prosper Rating (Alpha)')
plt.ylabel('Borrower Rate')
plt.title('Distribution of Borrower Rate and Prosper Rating')

plt.show();
```

## Visualization of Loan Status Distribution 

> This shows that Current is the highest loan status category in the dataset. 

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=[8,6])
sb.countplot(data=df_loan, x='LoanStatus',order = df_loan['LoanStatus'].value_counts().index, color=base_color)
plt.xlabel('Loan status ')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.title('Loan Status Distribution')
plt.show();
```

## Visualization of BorrowerAPR, ProsperRating (Alpha), Loan Terms 

> This visualization shows the BorrowerAPR decrease with the increasing loan terms for people with HR through to C ratings. \
> But this trend changes from prosperRating (Alpha) B through AA with longer loan terms (60 months) being higher.

```python
fig = plt.figure(figsize = [10, 6])
ax = sb.pointplot(data = df_loan, x = 'ProsperRating (Alpha)', y = 'BorrowerAPR', hue = 'Term',
           palette = 'crest', linestyles = '', dodge =0.2, errorbar='sd')
plt.title('Distribution of BorrowerAPR, ProsperRating (Alpha) and Term')
plt.ylabel('Mean Borrower APR')
ax.set_yticklabels([],minor = True);
```
